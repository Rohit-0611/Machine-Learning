Importing the Dataset

```python
titanic=sns.load_dataset("Titanic")
titanic
titanic=pd.DataFrame(titanic)
titanic
```

```python
titanic.drop(["alive","deck","embark_town","sibsp","parch","who","adult_male"],axis=1,inplace = True)
titanic
```

Splitting Dependent and Independent Columns

```python
x=titanic.iloc[:,1:-1].values
y=titanic.iloc[:,0].values
```

Replace Null values with Mean values

```python
from sklearn.impute import SimpleImputer
imuputer=SimpleImputer(missing_values=np.nan,strategy="mean")

#imputer made,apply it to age column

imuputer.fit(x[:,3:4])
x[:,3:4]=imuputer.transform(x[:,3:4])


```
