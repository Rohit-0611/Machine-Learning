```python
data=pd.read_csv("NFLX.csv")
data
```

Replacing the 0 values in volume with Average value

```python
#splitting the table
X=data.iloc[:,:-1].values
Y=data[["Volume"]].values
Y.shape
Y=np.array(Y)
Y
```

Replacing the null values with mean
